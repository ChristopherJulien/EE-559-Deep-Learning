# 3 Flops per second
Generate two square matrices of dimension 5000 x 5000 filled with random Gaussian coefficients,
compute their product, measure the time it takes, and estimate how many 
floating point products
have been executed per second (should be in the billions or tens of billions).
Hint: Use torch.empty, torch.normal_, torch.mm, and time.perf_counter.

```python
# Python program to show time by perf_counter()  
from time import perf_counter 

# Generate two square matrices of 5000x5000
m1 = torch.empty(5000,5000).normal_(mean=4,std=0.5)
m2 = torch.empty(5000,5000).normal_(mean=4,std=0.5)

# Start the stopwatch / counter 
t1_start = perf_counter()  

# compute their product
product = m1*m2

# Stop the stopwatch / counter 
t1_stop = perf_counter() 
t_duration = t1_stop-t1_start
  
print("Elapsed time during the whole program in seconds:", t_duration ) 

# Estimate the number of floating point products have beend executed per second
prod_per_sec = (5000^3 ) / (t_duration ) #! the number of floating point operations in nxn time n#n is equal to n^3
print("Estimate the number of floating point products have beend executed per second", prod_per_sec)
```

# 4 Playing with strides
Write a function mul_row, using python loops (and not even slicing operators), that gets a 2d tensor
as argument, and returns a tensor of same size, whose first row is identical to the first row of the
argument tensor, the second row is multiplied by two, the third by three, etc.

For instance:
>>> m = torch.full((4, 8), 2.0)
>>> m
tensor([[2., 2., 2., 2., 2., 2., 2., 2.],
[2., 2., 2., 2., 2., 2., 2., 2.],
[2., 2., 2., 2., 2., 2., 2., 2.],
[2., 2., 2., 2., 2., 2., 2., 2.]])
>>> mul_row(m)

tensor([[2., 2., 2., 2., 2., 2., 2., 2.],
[4., 4., 4., 4., 4., 4., 4., 4.],
[6., 6., 6., 6., 6., 6., 6., 6.],
[8., 8., 8., 8., 8., 8., 8., 8.]])


Then, write a second version named mul_row_fast, using tensor operations.
Apply both versions to a matrix of size 1000 x 400 and measure the time each takes (there should
be more than two orders of magnitude difference).

Hint: Use broadcasting and torch.arange, torch.view, torch.mul, and time.perf_counter.

```python
# Method with loop
def mul_row(arg_tensor):
    t2_start = perf_counter()  
    
    m = torch.empty(arg_tensor.size())
    rows = arg_tensor.size(0)
    for r in range(rows):
        m[r,:] = torch.mul(arg_tensor[r,:],(r+1))
    
    t2_stop = perf_counter()  
    print("Elapsed time during mul_row in seconds:", (t2_stop-t2_start) ) 

    return m
        
```

```python
# Method using tensor operations"
def mul_row_fast(arg_tensor):
    t3_start = perf_counter()  
    
    rows = arg_tensor.size(0)
    c = torch.arange(rows).add_(1) 
    mt = torch.transpose(arg_tensor,0,1)
    mul = torch.transpose(mt*c,0,1)
    
    t3_stop = perf_counter() 
    print("Elapsed time during mul_row_fast in seconds:", (t3_stop-t3_start) ) 
    
    return  mul
```

```python
m = torch.full((1000, 400), 1.0)

# m = torch.full((4, 8), 2.0)
m1 = mul_row(m)
m2 = mul_row_fast(m)

#! there should be more than two orders of magnitude difference

    
    
```
